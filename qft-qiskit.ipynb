## 1.1 Prep

```python
import numpy as np
from qiskit import QuantumCircuit
```

## 1.2 QFT

```python
def apply_qft(n, circuit):
    for j in range(n - 1, -1, -1):
        circuit.h(j)
        
        for k in range(j - 1, -1, -1):
            
            circuit.cp(np.pi / 2**(j - k), k, j)
        
        circuit.barrier()

    for i in range(n // 2):
        circuit.swap(i, n - i - 1)
```

```python
n3 = 3 
n4 = 4
circuit3 = QuantumCircuit (n3 ,0)
apply_qft (n3 , circuit3 )
print ( circuit3 )
print ()

circuit4 = QuantumCircuit (n4 ,0)
apply_qft (n4 , circuit4 )
print ( circuit4 )
print ()
```

## 1.3 Inverse QFT 

```python
def apply_iqft(n, circuit):
    for i in range(n // 2 - 1, -1, -1):
        circuit.swap(i, n - i - 1)

    for j in range(n):
        circuit.barrier()

        for k in range(j):
            shift = np.pi / 2**(j - k)
            circuit.cp(-shift, k, j)
            
        circuit.h(j)
```

```python
n = 4
circuit = QuantumCircuit (n ,0)
apply_iqft (n , circuit )
print ( circuit )
print ()
```

## 1.4 Generalizing the functions

```python
def apply_qft_gen(qubits, circuit):
    n = len(qubits)
    for j in range(n - 1, -1, -1):
        circuit.h(qubits[j])
        
        for k in range(j - 1, -1, -1):
            circuit.cp(np.pi / 2**(j - k), qubits[k], qubits[j])
        
        circuit.barrier()

    for i in range(n // 2):
        circuit.swap(qubits[i], qubits[n - i - 1])
```

```python
def apply_iqft_gen(qubits, circuit):
    n = len(qubits)
    for i in range(n // 2):
        circuit.swap(qubits[i], qubits[n - i - 1])

    for j in range(n):
        circuit.barrier()
        
        for k in range(j):
            circuit.cp(-np.pi / 2**(j - k), qubits[k], qubits[j])
            
        circuit.h(qubits[j])
```
